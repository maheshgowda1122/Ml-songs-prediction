```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/tracks.csv", engine='python')

print("\n📊 Sample Data:\n")
print(df[["energy","valence","popularity"]].head())

# Plot
plt.scatter(df["energy"], df["popularity"], alpha=0.3, label="Energy")
plt.scatter(df["valence"], df["popularity"], alpha=0.3, label="Valence")

plt.legend()
plt.xlabel("Feature Value")
plt.ylabel("Popularity")
plt.title("RQ3: Emotional Features vs Popularity")

plt.savefig("RQ3_emotional.pdf")
plt.show()
```
